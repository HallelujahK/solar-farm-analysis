***Import Imprtant Libraries***

```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

***Load the Datasets***

```python
data_benin = pd.read_csv("../data/data/benin-malanville.csv")
data_sierraleone = pd.read_csv("../data/data/sierraleone-bumbuna.csv")
data_togo = pd.read_csv("../data/data/togo-dapaong_qc.csv")
```

***Data Quality Check***

```python
print(data_benin.isnull().sum())
```

```python
sns.boxplot(data = data_benin[['GHI', 'DNI', 'DHI', 'ModA', 'ModB']])
plt.show()
```

### Time Series Analysis

```python
data_benin['Timestamp'] = pd.to_datetime(data_benin['Timestamp'])
plt.figure(figsize=(10, 6))
plt.plot(data_benin['Timestamp'], data_benin['GHI'], label='GHI')
plt.plot(data_benin['Timestamp'], data_benin['DNI'], label='DNI')
plt.plot(data_benin['Timestamp'], data_benin['DHI'], label='DHI')
plt.legend()
plt.show()
```

### Impact of Cleaning

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Cleaning', y='ModA', data=data_benin)
plt.show()
```

### Correlation Analysis

```python
plt.figure(figsize=(15, 6))
sns.heatmap(data_benin.corr(), annot=True, cmap='coolwarm')
plt.show()
```

### Wind Analysis

```python
sns.scatterplot(x='WD', y='WS', data=data_benin)
plt.show()
```

### Temperature Analysis

```python
sns.scatterplot(x='RH', y='Tamb', data=data_benin)
plt.show()
```

### Histograms

```python
data_benin[['GHI', 'DNI', 'DHI']].hist(bins=30, figsize=(10, 6))
plt.show()
```

### Z-Score Analysis

```python
from scipy.stats import zscore
data_benin['GHI_zscore'] = zscore(data_benin['GHI'])
outliers = data_benin[data_benin['GHI_zscore'].abs() > 3]
outliers.count()
```

### Bubble Charts

```python
plt.scatter(x=data_benin['GHI'], y=data_benin['Tamb'], s=data_benin['RH']*10)
plt.show()
```
