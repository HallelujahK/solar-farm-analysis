### Data Quality Check

```python
# Columns of interest
columns_of_interest = ['GHI', 'DNI', 'DHI', 'ModA', 'ModB', 'WS', 'WSgust']

# Function to clean negative values and replace outliers with NaN
def clean_outliers_and_negatives(df, column):
    # Replace negative values with NaN
    data_benin[column] = data_benin[column].apply(lambda x: np.nan if x < 0 else x)
    
    # Calculate Z-scores for the column
    z_scores = np.abs((data_benin[column] - data_benin[column].mean()) / data_benin[column].std())
    
    # Define an outlier threshold (e.g., z > 3)
    outlier_threshold = 3
    
    # Replace outliers with NaN
    data_benin[column] = np.where(z_scores > outlier_threshold, np.nan, data_benin[column])

# Apply the function to the columns of interest
for col in columns_of_interest:
    clean_outliers_and_negatives(data_benin, col)

for col in columns_of_interest:
    clean_outliers_and_negatives(data_sierraleone, col)

for col in columns_of_interest:
    clean_outliers_and_negatives(data_togo, col)

# Replace NaN values (including missing and cleaned outliers) with the mean of the column
data_benin[columns_of_interest] = data_benin[columns_of_interest].apply(lambda x: x.fillna(x.mean()))
data_sierraleone[columns_of_interest] = data_sierraleone[columns_of_interest].apply(lambda x: x.fillna(x.mean()))
data_togo[columns_of_interest] = data_togo[columns_of_interest].apply(lambda x: x.fillna(x.mean()))


# Check if the cleaning process was successful
data_benin[columns_of_interest].describe()
```
