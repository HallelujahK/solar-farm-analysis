***Data Quality Check***

```python
# Columns of interest
columns_of_interest = ['GHI', 'DNI', 'DHI', 'ModA', 'ModB', 'WS', 'WSgust']

# Function to clean negative values and replace outliers with NaN
def clean_outliers_and_negatives(df, column):
    # Replace negative values with NaN
    df[column] = df[column].apply(lambda x: np.nan if x < 0 else x)
    
    # Calculate Z-scores for the column
    z_scores = np.abs((df[column] - df[column].mean()) / df[column].std())
    
    # Define an outlier threshold (e.g., z > 3)
    outlier_threshold = 3
    
    # Replace outliers with NaN
    df[column] = np.where(z_scores > outlier_threshold, np.nan, df[column])

# Apply the function to the columns of interest
for col in columns_of_interest:
    clean_outliers_and_negatives(data_benin, col)

# Replace NaN values (including missing and cleaned outliers) with the mean of the column
data_benin[columns_of_interest] = data_benin[columns_of_interest].apply(lambda x: x.fillna(x.mean()))

# Check if the cleaning process was successful
data_benin[columns_of_interest].describe()
```

### Time Series Analysis

```python


# Convert the Timestamp column to datetime if it's not already
data_benin['Timestamp'] = pd.to_datetime(data_benin['Timestamp'])

# Set the Timestamp as the index
data_benin.set_index('Timestamp', inplace=True)

# Plotting GHI, DNI, DHI, and Tamb as line plots over time
plt.figure(figsize=(14, 8))

plt.plot(data_benin.index, data_benin['GHI'], label='GHI', color='blue')
plt.plot(data_benin.index, data_benin['DNI'], label='DNI', color='orange')
plt.plot(data_benin.index, data_benin['DHI'], label='DHI', color='green')
plt.plot(data_benin.index, data_benin['Tamb'], label='Tamb', color='red')

plt.xlabel('Time')
plt.ylabel('Values')
plt.title('GHI, DNI, DHI, and Tamb over Time')
plt.legend(loc='upper right')
plt.grid(True)

plt.show()

```

```python
# Plotting GHI, DNI, DHI, and Tamb as area plots over time
plt.figure(figsize=(14, 8))

plt.stackplot(data_benin.index, 
              data_benin['GHI'], 
              data_benin['DNI'], 
              data_benin['DHI'], 
              data_benin['Tamb'], 
              labels=['GHI', 'DNI', 'DHI', 'Tamb'],
              colors=['blue', 'orange', 'green', 'red'],
              alpha=0.6)

plt.xlabel('Time')
plt.ylabel('Values')
plt.title('GHI, DNI, DHI, and Tamb - Area Plot Over Time')
plt.legend(loc='upper right')
plt.grid(True)

plt.show()
```
